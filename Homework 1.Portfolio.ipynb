```python
start = dt.datetime(2012,3,6)
end = dt.datetime(2022,3,6)
```

```python
def get_stock(ticker):
    data=pdr.DataReader(f"{ticker}","yahoo",start,end)
    data[f"{ticker}"]=data["Adj Close"]
    data=data[[f'{ticker}']]
    print(data.head())
    return data
```

```python
apple = get_stock("AAPL")
samsung = get_stock("005930.KS")
hon = get_stock("HNHPF")
microsoft = get_stock("MSFT")
dell = get_stock("DELL")
sony = get_stock("SNEJF")
ibm = get_stock("IBM")
intel = get_stock("INTC")
pansonic = get_stock("PCRFY")
hp = get_stock("HPQ")
```

```python
def merge_stocks(tickers):
    data_frame = []
    for i in tickers:
            data_frame.append(get_stock(i))
    df_merged = reduce(lambda left, right:pd.merge(left,right,on=["Date"],how="outer"),data_frame)
    print(df_merged.head())
    return df_merged
```

```python
techstocks = ['AAPL','005930.KS','MSFT','DELL','SNEJF','IBM','INTC','PCRFY','HPQ']
```

```python
merge_stocks(techstocks)
```

```python
techport = merge_stocks(techstocks)
```

```python
techport.shape
```

```python
from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import CovarianceShrinkage
```

```python
mean = mean_historical_return(techport)
s = CovarianceShrinkage(techport).ledoit_wolf()
```

```python
from pypfopt.efficient_frontier import EfficientFrontier
```

```python
EF = EfficientFrontier(mean,s)
weights = EF.max_sharpe()
```

```python
clean_weights = EF.clean_weights()
print(dict(clean_weights))
```

```python
EF.portfolio_performance(verbose=True)
```

```python
from pypfopt.discrete_allocation import DiscreteAllocation,get_latest_prices
```

```python
new_prices = get_latest_prices(techport)
DA = DiscreteAllocation(weights,new_prices,total_portfolio_value=10000)

allocation,leftover=DA.greedy_portfolio()

print("Discrete Allocation:",allocation)
print("Funds Remaining: ${:.2f}".format(leftover))
```
